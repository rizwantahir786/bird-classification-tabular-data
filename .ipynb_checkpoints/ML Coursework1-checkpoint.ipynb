## Removing Outliers Using Inter-Quartile Range (IQR) 

```python
df
```

```python
Features = ['huml','humw','feml','femw','tibl','tarl','tarw']
for feature in Features:
    # Finding the IQR
    percentile25 = df[feature].quantile(0.25)
    percentile75 = df[feature].quantile(0.75)
    iqr= percentile75 - percentile25  
    # Finding Upper and Lower Limit
    upper_limit = percentile75 + 1.5 *iqr
    lower_limit = percentile25 - 1.5 *iqr
    # removing outleirs
    df = df.loc[(df[feature] > lower_limit) & (df[feature] < upper_limit)]    
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.metrics import confusion_matrix, classification_report
import scikitplot as skplt
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)
#Prediction
rfc_prediction = rfc.predict(X_test)
cm = confusion_matrix(y_test,rfc_prediction)

#Evaluation
print('Classification Report')
print(classification_report(y_test,rfc_prediction))

print('Confusion Matrix')
skplt.metrics.plot_confusion_matrix(y_test,rfc_prediction,figsize=(8,8))
```

```python
def Specificity(cm):
    class1= (cm[1,1]+cm[2,2]+cm[3,3]+cm[4,4]+cm[5,5])/(cm[1,0]+cm[2,0]+cm[3,0]+cm[4,0]+cm[5,0]+cm[1,1]+cm[2,2]+cm[3,3]+cm[4,4]+cm[5,5])
    class2= (cm[0,0]+cm[2,2]+cm[3,3]+cm[4,4]+cm[5,5])/(cm[0,1]+cm[2,1]+cm[3,1]+cm[4,1]+cm[5,1]+cm[0,0]+cm[2,2]+cm[3,3]+cm[4,4]+cm[5,5])
    class3= (cm[0,0]+cm[1,1]+cm[3,3]+cm[4,4]+cm[5,5])/(cm[0,2]+cm[1,2]+cm[3,2]+cm[4,2]+cm[5,2]+cm[0,0]+cm[1,1]+cm[3,3]+cm[4,4]+cm[5,5])
    class4= (cm[0,0]+cm[1,1]+cm[2,2]+cm[4,4]+cm[5,5])/(cm[0,3]+cm[1,3]+cm[2,3]+cm[4,3]+cm[5,3]+cm[0,0]+cm[1,1]+cm[2,2]+cm[4,4]+cm[5,5])
    class5= (cm[0,0]+cm[1,1]+cm[2,2]+cm[3,3]+cm[5,5])/(cm[0,4]+cm[1,4]+cm[2,4]+cm[3,4]+cm[5,4]+cm[0,0]+cm[1,1]+cm[2,2]+cm[3,3]+cm[5,5])
    class6= (cm[0,0]+cm[1,1]+cm[2,2]+cm[3,3]+cm[4,4])/(cm[0,5]+cm[1,5]+cm[2,5]+cm[3,5]+cm[4,5]+cm[0,0]+cm[1,1]+cm[2,2]+cm[3,3]+cm[4,4])
    List = [class1,class2,class3,class4,class5,class6]
    return List

def Sensitivity(cm):
    class1=cm[0,0]/(cm[0,0]+cm[0,1]+cm[0,2]+cm[0,3]+cm[0,4]+cm[0,5])
    class2=cm[1,1]/(cm[1,0]+cm[1,1]+cm[1,2]+cm[1,3]+cm[1,4]+cm[1,5])
    class3=cm[2,2]/(cm[2,0]+cm[2,1]+cm[2,2]+cm[2,3]+cm[2,4]+cm[2,5])
    class4=cm[3,3]/(cm[3,0]+cm[3,1]+cm[3,2]+cm[3,3]+cm[3,4]+cm[3,5])
    class5=cm[4,4]/(cm[4,0]+cm[4,1]+cm[4,2]+cm[4,3]+cm[4,4]+cm[4,5])
    class6=cm[5,5]/(cm[5,0]+cm[5,1]+cm[5,2]+cm[5,3]+cm[5,4]+cm[5,5])
    List = [class1,class2,class3,class4,class5,class6]
    return List

def Accuracy(cm):
    class1=(cm[0,0]+cm[1,1]+cm[2,2]+cm[3,3]+cm[4,4]+cm[5,5])/(cm[0,0]+cm[0,1]+cm[0,2]+cm[0,3]+cm[0,4]+cm[0,5]+cm[1,0]+cm[2,0]+cm[3,0]+cm[4,0]+cm[5,0]+cm[1,1]+cm[2,2]+cm[3,3]+cm[4,4]+cm[5,5])
    class2=(cm[1,1]+cm[0,0]+cm[2,2]+cm[3,3]+cm[4,4]+cm[5,5])/(cm[1,0]+cm[1,1]+cm[1,2]+cm[1,3]+cm[1,4]+cm[1,5]+cm[0,1]+cm[2,1]+cm[3,1]+cm[4,1]+cm[5,1]+cm[0,0]+cm[2,2]+cm[3,3]+cm[4,4]+cm[5,5])
    class3=(cm[2,2]+cm[0,0]+cm[1,1]+cm[3,3]+cm[4,4]+cm[5,5])/(cm[2,0]+cm[2,1]+cm[2,2]+cm[2,3]+cm[2,4]+cm[2,5]+cm[0,2]+cm[1,2]+cm[3,2]+cm[4,2]+cm[5,2]+cm[0,0]+cm[1,1]+cm[3,3]+cm[4,4]+cm[5,5])
    class4=(cm[3,3]+cm[0,0]+cm[1,1]+cm[2,2]+cm[4,4]+cm[5,5])/(cm[3,0]+cm[3,1]+cm[3,2]+cm[3,3]+cm[3,4]+cm[3,5]+cm[0,3]+cm[1,3]+cm[2,3]+cm[4,3]+cm[5,3]+cm[0,0]+cm[1,1]+cm[2,2]+cm[4,4]+cm[5,5])
    class5=(cm[4,4]+cm[0,0]+cm[1,1]+cm[2,2]+cm[3,3]+cm[5,5])/(cm[4,0]+cm[4,1]+cm[4,2]+cm[4,3]+cm[4,4]+cm[4,5]+cm[0,4]+cm[1,4]+cm[2,4]+cm[3,4]+cm[5,4]+cm[0,0]+cm[1,1]+cm[2,2]+cm[3,3]+cm[5,5])
    class6=(cm[5,5]+cm[0,0]+cm[1,1]+cm[2,2]+cm[3,3]+cm[4,4])/(cm[5,0]+cm[5,1]+cm[5,2]+cm[5,3]+cm[5,4]+cm[5,5]+cm[0,5]+cm[1,5]+cm[2,5]+cm[3,5]+cm[4,5]+cm[0,0]+cm[1,1]+cm[2,2]+cm[3,3]+cm[4,4])
    List = [class1,class2,class3,class4,class5,class6]
    return List
```

```python
x=Specificity(cm)
print('Specificity = ',x)
y=Sensitivity(cm)
print('Sensitivity = ',y)
z=Accuracy(cm)
print('Accuracy = ',z)
```
